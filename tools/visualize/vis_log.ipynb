```python
import json
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d

def get_hard_sample(file_dir, name):
    f = open(file_dir, encoding="utf-8")
    lines = f.readlines()      # 
    elements = []
    for line in lines:
        if name in line:
            dicts = json.loads(line)
            elements.append(dicts[name])
    return elements

file_dir = '/home/linhuadong/DNFNet/jobs/prw_dnfnet2_max_iou_origin_cluster_sample/work_dirs/prw/20230305_222019.log.json'
global_cluster_hard_loss_sample = get_hard_sample(file_dir, name='global_cluster_hard_loss_sample')
global_cluster_hard_loss = get_hard_sample(file_dir, name='global_cluster_hard_loss')
plt.figure(figsize=(6, 6), dpi=100)
plt.grid(linestyle="--")  # 

plt.plot(range(len(global_cluster_hard_loss_sample)), global_cluster_hard_loss_sample, c='r', linewidth=2, label='sample')
plt.plot(range(len(global_cluster_hard_loss)), global_cluster_hard_loss, c='g', linewidth=2, label='origin')

plt.legend()

plt.xticks(fontsize=12)  # 10
plt.yticks(fontsize=12)
plt.xlabel("iterations", fontsize=12)
# plt.ylabel("global_cluster_hard_loss_sample", fontsize=12)

# plt.legend(loc=0, prop = {'size':12})
# plt.savefig('./filename.jpg', format='jpg', bbox_inches='tight')  # svg,inkscapeemfword
# plt.savefig('./file/outliers.eps', dpi=100, format='eps', bbox_inches='tight', pad_inches=0)
plt.show()
```
