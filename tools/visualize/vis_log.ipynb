```python
import json
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d

def get_hard_sample(file_dir):
    f = open(file_dir, encoding="utf-8")
    lines = f.readlines()      #读取全部内容 ，并以列表方式返回
    elements = []
    for line in lines:
        if "memory_similarity" in line:
            dicts = json.loads(line)
            elements.append(dicts['memory_similarity'])
    return elements

file_dir = '/home/linhuadong/DNFNet/jobs/prw_test_memory_sim/work_dirs/prw/20221219_180430.log.json'
elements = get_hard_sample(file_dir)

plt.figure(figsize=(6, 6), dpi=100)
plt.grid(linestyle="--")  # 设置背景网格线为虚线

plt.plot(range(len(elements)), elements, c='r', linewidth=2)

plt.xticks(fontsize=12)  # 默认字体大小为10
plt.yticks(fontsize=12)
plt.xlabel("iterations", fontsize=12)
plt.ylabel("similarity", fontsize=12)

# plt.legend(loc=0, prop = {'size':12})
# plt.savefig('./filename.jpg', format='jpg', bbox_inches='tight')  # 建议保存为svg格式,再用inkscape转为矢量图emf后插入word中
# plt.savefig('./file/outliers.eps', dpi=100, format='eps', bbox_inches='tight', pad_inches=0)
plt.show()
```
