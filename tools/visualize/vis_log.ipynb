```python
import json
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d

def get_hard_sample(file_dir):
    f = open(file_dir, encoding="utf-8")
    lines = f.readlines()      # 
    elements = []
    for line in lines:
        if "average_std_exclude_outliers" in line:
            dicts = json.loads(line)
            elements.append(dicts['average_std_exclude_outliers'])
    return elements

file_dir = '/home/linhuadong/DNFNet/jobs/prw_dnfnet_max_iou_update_memory_triplet2/work_dirs/prw/20230303_131218.log.json'
elements = get_hard_sample(file_dir)
print(elements)
plt.figure(figsize=(6, 6), dpi=100)
plt.grid(linestyle="--")  # 

plt.plot(range(len(elements)), elements, c='r', linewidth=2)

plt.xticks(fontsize=12)  # 10
plt.yticks(fontsize=12)
plt.xlabel("iterations", fontsize=12)
plt.ylabel("average_std_exclude_outliers", fontsize=12)

# plt.legend(loc=0, prop = {'size':12})
# plt.savefig('./filename.jpg', format='jpg', bbox_inches='tight')  # svg,inkscapeemfword
# plt.savefig('./file/outliers.eps', dpi=100, format='eps', bbox_inches='tight', pad_inches=0)
plt.show()
```
