# 对json文件中某些字段进行可视化

```python
import json
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d

def get_hard_sample(file_dir, name):
    f = open(file_dir, encoding="utf-8")
    lines = f.readlines()      #读取全部内容 ，并以列表方式返回
    elements = []
    for line in lines:
        if name in line:
            dicts = json.loads(line)
            elements.append(dicts[name])
    return elements

file_dir = '/home/linhuadong/DNFNet/jobs/prw_dnfnet2_nopart_uncertainty_centroid/work_dirs/prw/20230512_115423.log.json'
global_cluster_hard_loss_sample = get_hard_sample(file_dir, name='uncertainty sample number')
plt.figure(figsize=(6, 6), dpi=100)
plt.grid(linestyle="--")  # 设置背景网格线为虚线
plt.plot(range(len(global_cluster_hard_loss_sample)), global_cluster_hard_loss_sample, c='r', linewidth=2, label='sample')
plt.legend()

plt.xticks(fontsize=12)  # 默认字体大小为10
plt.yticks(fontsize=12)
plt.xlabel("iterations", fontsize=12)
# plt.ylabel("global_cluster_hard_loss_sample", fontsize=12)
# plt.legend(loc=0, prop = {'size':12})
# plt.savefig('./filename.jpg', format='jpg', bbox_inches='tight')  # 建议保存为svg格式,再用inkscape转为矢量图emf后插入word中
# plt.savefig('./file/outliers.eps', dpi=100, format='eps', bbox_inches='tight', pad_inches=0)
plt.show()
```

# 对log文件中某些字段进行可视化

### log文件

```python
import json
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d

def get_hard_sample(file_dir, name):
    f = open(file_dir, encoding="utf-8")
    lines = f.readlines()      #读取全部内容 ，并以列表方式返回
    elements = []
    for line in lines:
        if name in line:
            element = line.split(":")[-1][1:]
            elements.append(int(element))
    return elements

file_dir = '/home/linhuadong/DNFNet/jobs/prw_cpcl_all_tricks/work_dirs/prw/20230717_224558.log'
file_dir = '/home/linhuadong/DNFNet/jobs/prw_dnfnet2_all_trick_modify_finch/work_dirs/prw/20230716_190036.log'
global_cluster_hard_loss_sample = get_hard_sample(file_dir, name='uncertainty sample number')
plt.figure(figsize=(10, 6), dpi=100)
plt.grid(linestyle="--")  # 设置背景网格线为虚线
plt.plot(range(len(global_cluster_hard_loss_sample)), global_cluster_hard_loss_sample, c='r', linewidth=2, label='sample')
plt.legend()

plt.xticks(fontsize=12)  # 默认字体大小为10
plt.yticks(fontsize=12)
plt.xlabel("iterations", fontsize=12)
plt.ylabel("number of boundary samples", fontsize=12)

plt.legend(loc=0, prop = {'size':12})
plt.savefig('./boundary_samples.jpg', format='jpg', bbox_inches='tight')  # 建议保存为svg格式,再用inkscape转为矢量图emf后插入word中
plt.savefig('./file/boundary_samples.eps', dpi=100, format='eps', bbox_inches='tight', pad_inches=0)
plt.show()
```
