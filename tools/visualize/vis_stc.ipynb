```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Arial']  # ,SimHei
plt.rcParams['axes.unicode_minus'] = False  # 
colors = ['b', 'g','maroon', 'c', 'm', 'y', 'k', 'r']
markers = ['o', 'v', '^', 'D', '>', 's', '*', '+']
linestyles = ['-', '-', '-', '-', '-', '--', '--', '--']
(x, y) = (0, 1)
x_values = [x for x in range(0, 800)]
plt.figure(figsize=(10, 5), dpi=100)
ms = [-0.0005, -0.001, -0.0015, -0.002]
plt.grid(linestyle="--")  # 
for (i, m) in enumerate(ms):
    b = y - m * x
    y_values = [m * x + b if m * x + b > 0 else 0 for x in x_values]
    plt.plot(x_values, y_values, label=f'y = {m}x + {b}', linewidth=2)
    
    # plt.plot(x, methods[name], marker=markers[i], markersize=5, color=colors[i], label=name, linewidth=2, linestyle=linestyles[i])

# group_labels = [50, 100, 500, 1000, 2000, 4000]  # x
# plt.xticks(x, group_labels, fontsize=12)  # 10
# plt.ylim((40, 100))
# plt.title('f(dT)=k * dT + 1')  # 
plt.xlabel("dT", fontsize=12)
plt.ylabel("f(dT)", fontsize=12)

plt.legend(loc=0, prop = {'size':12})
# plt.savefig('../file/gallery_size.eps', dpi=100, format='eps', bbox_inches='tight', pad_inches=0)
plt.show()
```
